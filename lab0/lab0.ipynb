## 2. Classes and Objects

### Object and Class
In python, every python value is an object, and every object has a type. You can define your own classes to create custom objects with specific attributes and behaviors. A class is a blueprint for creating objects, and an object is an instance of a class. A class is a blueprint for creating objects, and an object is an instance of a class. Classes can have attributes (data) and methods (functions that operate on the data).

> Why we need classes and objects?
> Imagine that we have several related AI models in our project, modelA, modelB, and modelC. Each model has its own parameters and methods for training and inference. Instead of writing separate code for each model, we can define a class `nn.Module` that encapsulates the common attributes and behaviors of all models. Then, we can create instances of this class for each specific model, allowing us to reuse code and maintain a cleaner structure.

````python
import math

class Vector:
    """A simple 2D vector class with magic methods."""
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        """String representation for print()."""
        return f"Vector({self.x}, {self.y})"
    
    def __repr__(self):
        """Official string representation."""
        return f"Vector(x={self.x}, y={self.y})"
    
    def __add__(self, other):
        """Define addition: v1 + v2."""
        return Vector(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):
        """Define subtraction: v1 - v2."""
        return Vector(self.x - other.x, self.y - other.y)
    
    def __mul__(self, scalar):
        """Define scalar multiplication: v * scalar."""
        return Vector(self.x * scalar, self.y * scalar)
    
    def __eq__(self, other):
        """Define equality: v1 == v2."""
        return self.x == other.x and self.y == other.y
    
    def __len__(self):
        """Define length (magnitude)."""
        return int((self.x**2 + self.y**2)**0.5)
    
    def __getitem__(self, index):
        """Enable indexing: v[0], v[1]."""
        if index == 0:
            return self.x
        elif index == 1:
            return self.y
        else:
            raise IndexError("Vector index out of range")

# Test magic methods
v1 = Vector(3, 4)
v2 = Vector(1, 2)

print(f"v1: {v1}")           # __str__
print(f"v2: {v2}")
print(f"v1 + v2: {v1 + v2}") # __add__
print(f"v1 - v2: {v1 - v2}") # __sub__
print(f"v1 * 3: {v1 * 3}")   # __mul__
print(f"v1 == v2: {v1 == v2}")  # __eq__
print(f"len(v1): {len(v1)}")    # __len__
print(f"v1[0]: {v1[0]}, v1[1]: {v1[1]}")  # __getitem__

# Or you can access directly
print(f"v1.x: {v1.x}, v1.y: {v1.y}")
````

### OOP briefly

OOP — Classes, Dunder Methods & Inheritance

**Why OOP now?** Every PyTorch model is defined by subclassing `torch.nn.Module`:

```python
import torch.nn as nn
class MyModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc = nn.Linear(4, 2)
    def forward(self, x):
        return self.fc(x)
```

Why we use inheritance here? In programming, we don't want to reinvent the wheel every time we create a new model. By inheriting from `nn.Module`, we can reuse the functionality provided by PyTorch, such as parameter management, saving/loading models, and more. This allows us to focus on defining the unique aspects of our model while leveraging the powerful features of the base class. Inheritance allows us to create a new class that is a modified version of an existing class. By inheriting from `nn.Module`, we can take advantage of the built-in methods and properties that PyTorch provides for model management, such as handling parameters, saving/loading models, and more. This promotes code reuse and makes it easier to build complex models without having to implement all the underlying functionality from scratch.

We build a minimal stand-in today so this pattern is fully familiar when we reach Lab 6.

### Key dunder methods

| Method | Triggered by | PyTorch analogy |
|--------|--------------|-----------------|
| `__init__` | `MyClass(...)` | layer definitions |
| `__repr__` | `repr(obj)` / REPL | model summary string |
| `__call__` | `obj(x)` | dispatches to `forward(x)` |
| `__len__` | `len(obj)` | `Dataset.__len__` |
| `__getitem__` | `obj[i]` | `Dataset.__getitem__` |
| `__add__` | `a + b` | tensor operator overloading |

````python
# ── Minimal Module base class — mirrors torch.nn.Module ───────────────────────
# Note that, you don't need to fully understand the behavior of this code right now. We will cover it in detail in the next lecture. Just taste the code.

class Module:
    """Base class. __call__ dispatches to forward() so callers write
    model(x), not model.forward(x) — exactly like real PyTorch.
    """
    def __call__(self, x):
        return self.forward(x)

    def forward(self, x):
        raise NotImplementedError("Subclasses must implement forward()")

    def __repr__(self):
        return self.__class__.__name__ + "()"


class ReLU(Module):
    """max(0, x) — most common hidden-layer activation."""
    def forward(self, x):
        return max(0.0, x)


class Sigmoid(Module):
    """1 / (1 + exp(-x)) — output activation for binary classification."""
    def forward(self, x):
        import math
        return 1.0 / (1.0 + math.exp(-x))


class Sequential(Module):
    """Chains modules in order. Sequential(ReLU(), Sigmoid())(x) applies both."""
    def __init__(self, *layers):
        self.layers = layers

    def forward(self, x):
        for layer in self.layers:
            x = layer(x)
        return x

    def __repr__(self):
        return "Sequential(" + ", ".join(repr(l) for l in self.layers) + ")"


relu    = ReLU()
sigmoid = Sigmoid()
model   = Sequential(ReLU(), Sigmoid())

for val in [-2.0, 0.0, 2.0]:
    print(f"x={val:+.1f}  relu={relu(val):.4f}  sigmoid={sigmoid(val):.4f}  seq={model(val):.4f}")

print("\nmodel repr :", model)
print("inheritance:", issubclass(Sequential, Module))
````

### Copy object

Common mistake: `copy_sth = original_sth` does not create a new object, it just creates a new reference to the same object. 
In other words, `copy_sth` and `original_sth` are just two names for the same object in memory. If you modify `copy_sth`, you will also modify `original_sth`, because they refer to the same object. To create a true copy of an object, you can use the `copy` module from the Python standard library, which provides functions for shallow and deep copying of objects. For example, `copy.copy(original_sth)` creates a shallow copy of the object, while `copy.deepcopy(original_sth)` creates a deep copy that recursively copies all nested objects as well. Always be mindful of whether you are creating a new object or just a new reference to an existing object when working with variables in Python.

````python
# Assignment creates a reference, NOT a copy
list1 = [1, 2, 3, [4, 5]]
list2 = list1  # This is just a reference!

print("Before modification:")
print(f"list1: {list1}")
print(f"list2: {list2}")
print(f"list1 is list2: {list1 is list2}")

# id is the location of the object in memory. Since list1 and list2 refer to the same object, they have the same id.
print(f"id(list1): {id(list1)}")
print(f"id(list2): {id(list2)}")

# Modify list2
list2[0] = 999

print("\nAfter modifying list2[0]:")
print(f"list1: {list1}")  # list1 also changed!
print(f"list2: {list2}")
````

````python
import copy

# Shallow copy: creates a new object, but nested objects are still references
list1 = [1, 2, 3, [4, 5]]
list2 = list1.copy()  # or copy.copy(list1) or list1[:]

print("After shallow copy:")
print(f"list1: {list1}")
print(f"list2: {list2}")
print(f"list1 is list2: {list1 is list2}")

# Modify top-level element
list2[0] = 999
print("\nAfter modifying list2[0] (top-level):")
print(f"list1: {list1}")  # list1 NOT affected
print(f"list2: {list2}")

# Modify nested element
list2[3][0] = 888
print("\nAfter modifying list2[3][0] (nested):")
print(f"list1: {list1}")  # list1 IS affected! (shallow copy problem)
print(f"list2: {list2}")
````

````python
import copy

# Deep copy: creates a completely independent copy
list1 = [1, 2, 3, [4, 5]]
list2 = copy.deepcopy(list1)

print("After deep copy:")
print(f"list1: {list1}")
print(f"list2: {list2}")

# Modify nested element
list2[3][0] = 888
print("\nAfter modifying list2[3][0] (nested):")
print(f"list1: {list1}")  # list1 NOT affected!
print(f"list2: {list2}")
````

### Assert

The `assert` statement is used to check if a condition is true. If the condition is false, it raises an `AssertionError` with an optional error message. This is useful for debugging and ensuring that certain conditions hold true in your code. For example:
```python
x = 10
assert x > 0, "x must be positive"
```
In this example, if `x` is not greater than 0, an `AssertionError` will be raised with the message "x must be positive". This helps you catch bugs early and ensure that your code is behaving as expected.

`try-except` blocks are used to handle exceptions in Python. They allow you to catch and handle errors gracefully without crashing your program. For example:
```python
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f"Error: {e}")
```
In this example, we are trying to divide 10 by 0, which will raise a `ZeroDivisionError`. By using a `try-except` block, we can catch this error and print a custom error message instead of crashing the program.

````python
def add(a, b):
    return a + b

# Test with assertions
assert add(2, 3) == 5, "2 + 3 should be 5"
assert add(-1, 1) == 0, "-1 + 1 should be 0"
assert add(0, 0) == 0, "0 + 0 should be 0"

print("PASSED!")

# This will fail
try:
    assert add(2, 2) == 5, "This assertion will fail"
except AssertionError as e:
    print(f"✗ Test failed: {e}")
````

### Profiling and Debugging in Python

Profiling and debugging are essential skills for any programmer, especially when working with complex AI models. Python provides several tools for profiling and debugging your code.
- `cProfile`: A built-in profiler that provides a comprehensive report of the time spent in each function.
- `line_profiler`: A third-party library that allows you to profile the time spent on each line of code within a function.
- `pdb`: The built-in Python debugger that allows you to step through your code, inspect variables, and evaluate expressions interactively.

````python
import cProfile, pstats, io as _io, random, time

# ── Two implementations ───────────────────────────────────────────────────────
def py_matmul(A, B):
    """Pure-Python O(n^3) — has Python interpreter overhead on every op."""
    n = len(A)
    C = [[0.0]*n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C

def np_matmul(A, B):
    """NumPy — BLAS-backed, no Python overhead per element."""
    return A @ B

# Prepare small test data (N=40 keeps cProfile run short)
random.seed(0); N = 40
A_list = [[random.gauss(0,1) for _ in range(N)] for _ in range(N)]
B_list = [[random.gauss(0,1) for _ in range(N)] for _ in range(N)]

print(f"Test matrix size: {N}×{N}")
````

````python
# ── cProfile via Python API ───────────────────────────────────────────────────
pr = cProfile.Profile()
pr.enable()
py_matmul(A_list, B_list)
pr.disable()

buf = _io.StringIO()
pstats.Stats(pr, stream=buf).sort_stats('tottime').print_stats(8)
print(buf.getvalue())
````

````python
# ── Shape-assertion guard — used in every forward() from Lab 6 onward ─────────
def checked_matmul(A: np.ndarray, B: np.ndarray) -> np.ndarray:
    assert A.ndim == 2 and B.ndim == 2, "Both inputs must be 2-D"
    assert A.shape[1] == B.shape[0], f"Shape mismatch: {A.shape} @ {B.shape}"
    return A @ B

C = checked_matmul(np.ones((3,4)), np.ones((4,2)))
print("Result shape:", C.shape)   # (3, 2)
````
