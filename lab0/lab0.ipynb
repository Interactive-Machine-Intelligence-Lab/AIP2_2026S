## Iterables and Generators

### Iterables

An **iterable** is any Python object capable of returning its members one at a time, allowing it to be iterated over in a loop. Common examples include lists, tuples, and strings.

```python
# Example of an iterable

    print(item)
```

### Generators

A **generator** is a special type of iterable that generates values on-the-fly and does not store the entire sequence in memory. Generators are defined using functions and the `yield` keyword.

```python
# Example of a generator function
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

# Using the generator
for number in count_up_to(5):
    print(number)
```

## Section 5 · OOP — Classes, Dunder Methods & Inheritance

**Why OOP now?** Every PyTorch model is defined by subclassing `torch.nn.Module`:

```python
import torch.nn as nn
class MyModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc = nn.Linear(4, 2)
    def forward(self, x):
        return self.fc(x)
```

Why we use inheritance here? In programming, we don't want to reinvent the wheel every time we create a new model. By inheriting from `nn.Module`, we can reuse the functionality provided by PyTorch, such as parameter management, saving/loading models, and more. This allows us to focus on defining the unique aspects of our model while leveraging the powerful features of the base class. Inheritance allows us to create a new class that is a modified version of an existing class. By inheriting from `nn.Module`, we can take advantage of the built-in methods and properties that PyTorch provides for model management, such as handling parameters, saving/loading models, and more. This promotes code reuse and makes it easier to build complex models without having to implement all the underlying functionality from scratch.

We build a minimal stand-in today so this pattern is fully familiar when we reach Lab 6.

### Key dunder methods

| Method | Triggered by | PyTorch analogy |
|--------|--------------|-----------------|
| `__init__` | `MyClass(...)` | layer definitions |
| `__repr__` | `repr(obj)` / REPL | model summary string |
| `__call__` | `obj(x)` | dispatches to `forward(x)` |
| `__len__` | `len(obj)` | `Dataset.__len__` |
| `__getitem__` | `obj[i]` | `Dataset.__getitem__` |
| `__add__` | `a + b` | tensor operator overloading |

````python
# ── Minimal Module base class — mirrors torch.nn.Module ───────────────────────

class Module:
    """Base class. __call__ dispatches to forward() so callers write
    model(x), not model.forward(x) — exactly like real PyTorch.
    """
    def __call__(self, x):
        return self.forward(x)

    def forward(self, x):
        raise NotImplementedError("Subclasses must implement forward()")

    def __repr__(self):
        return self.__class__.__name__ + "()"


class ReLU(Module):
    """max(0, x) — most common hidden-layer activation."""
    def forward(self, x):
        return max(0.0, x)


class Sigmoid(Module):
    """1 / (1 + exp(-x)) — output activation for binary classification."""
    def forward(self, x):
        import math
        return 1.0 / (1.0 + math.exp(-x))


class Sequential(Module):
    """Chains modules in order. Sequential(ReLU(), Sigmoid())(x) applies both."""
    def __init__(self, *layers):
        self.layers = layers

    def forward(self, x):
        for layer in self.layers:
            x = layer(x)
        return x

    def __repr__(self):
        return "Sequential(" + ", ".join(repr(l) for l in self.layers) + ")"


relu    = ReLU()
sigmoid = Sigmoid()
model   = Sequential(ReLU(), Sigmoid())

for val in [-2.0, 0.0, 2.0]:
    print(f"x={val:+.1f}  relu={relu(val):.4f}  sigmoid={sigmoid(val):.4f}  seq={model(val):.4f}")

print("\nmodel repr :", model)
print("inheritance:", issubclass(Sequential, Module))
````

# Profiling: `cProfile` and `line_profiler`

ML training loops run millions of iterations — performance matters.

| Tool | Granularity | When to use |
|------|-------------|-------------|
| `cProfile` / `%prun` | per **function** | Find which function is the bottleneck |
| `line_profiler` / `%lprun` | per **line** | Understand *why* a specific function is slow |

**Workflow:** `cProfile` first → identify hot function → `line_profiler` to zoom in.

````python
import cProfile, pstats, io as _io, random, time

# ── Two implementations ───────────────────────────────────────────────────────
def py_matmul(A, B):
    """Pure-Python O(n^3) — has Python interpreter overhead on every op."""
    n = len(A)
    C = [[0.0]*n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C

def np_matmul(A, B):
    """NumPy — BLAS-backed, no Python overhead per element."""
    return A @ B

# Prepare small test data (N=40 keeps cProfile run short)
random.seed(0); N = 40
A_list = [[random.gauss(0,1) for _ in range(N)] for _ in range(N)]
B_list = [[random.gauss(0,1) for _ in range(N)] for _ in range(N)]

print(f"Test matrix size: {N}×{N}")
````

````python
# ── cProfile via Python API ───────────────────────────────────────────────────
pr = cProfile.Profile()
pr.enable()
py_matmul(A_list, B_list)
pr.disable()

buf = _io.StringIO()
pstats.Stats(pr, stream=buf).sort_stats('tottime').print_stats(8)
print(buf.getvalue())
````

````python
# ── %prun — one-liner cProfile magic ─────────────────────────────────────────
%prun -l 6 py_matmul(A_list, B_list)
````

````python
# ── line_profiler — line-by-line timing ───────────────────────────────────────
# install: uv add line-profiler
%load_ext line_profiler
%lprun -f py_matmul py_matmul(A_list, B_list)
````

````python
# ── Shape-assertion guard — used in every forward() from Lab 6 onward ─────────
def checked_matmul(A: np.ndarray, B: np.ndarray) -> np.ndarray:
    assert A.ndim == 2 and B.ndim == 2, "Both inputs must be 2-D"
    assert A.shape[1] == B.shape[0], f"Shape mismatch: {A.shape} @ {B.shape}"
    return A @ B

C = checked_matmul(np.ones((3,4)), np.ones((4,2)))
print("Result shape:", C.shape)   # (3, 2)
````

# NumPy Basics

Numpy is a fundamental library for numerical computing in Python. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays efficiently. NumPy is widely used in data science, machine learning, and scientific computing due to its performance and ease of use.

Numpy has own classes as well, such as `numpy.ndarray`, which is the core data structure for storing and manipulating arrays. It provides a powerful interface for performing mathematical operations on arrays, including element-wise operations, broadcasting, and linear algebra functions. NumPy also includes a wide range of functions for creating arrays, such as `numpy.array()`, `numpy.zeros()`, `numpy.ones()`, and more. These functions allow you to easily create arrays of different shapes and sizes to suit your needs.

The main difference between NumPy arrays and Python lists is that NumPy arrays are more efficient for numerical computations and support a wider range of mathematical operations. NumPy arrays are implemented in C and optimized for performance, while Python lists are more general-purpose and can hold heterogeneous data types. Additionally, NumPy arrays support broadcasting, which allows for operations to be performed on arrays of different shapes without the need for explicit loops, making code more concise and efficient. 

For further reading, check out the [NumPy documentation](https://numpy.org/doc/stable/).

````python
# ── Broadcasting ─────────────────────────────────────────────────────────────
# NumPy stretches size-1 dimensions to match, without copying data.
#
# Rules (right-aligned shapes):
#   A (3, 3) + v (3,) → v is treated as (1, 3) → broadcast to (3, 3)

A = np.ones((3, 3))
v = np.array([1.0, 2.0, 3.0])    # shape (3,)
col = np.array([[10.0],[20.0],[30.0]])  # shape (3,1)

print("A + v  (row broadcast):\n", A + v)
print("A + col (col broadcast):\n", A + col)
print("outer product v @ col.T:\n", v.reshape(-1,1) @ col.reshape(1,-1))
````

# Vectorize

Now, we have `np.array` for fast computing, we may curious about its performance. Let's see how it compares to a Python list for a simple operation like element-wise addition.

````python
def measure(fn, repeats=7):
    """Return minimum wall-clock time over `repeats` runs (seconds)."""
    times = []
    for _ in range(repeats):
        t0 = time.perf_counter()
        fn()
        times.append(time.perf_counter() - t0)
    return min(times)

N = 1000
np.random.seed(42)

v_np = np.random.randn(N);       v_l = v_np.tolist()
w_np = np.random.randn(N);       w_l = w_np.tolist()
A_np = np.random.randn(N, N);    A_l = A_np.tolist()

# ── 1. Element-wise square ────────────────────────────────────────────────────
sp_sq = (measure(lambda: [x*x for x in v_l]) /
         measure(lambda: v_np**2))

# ── 2. Dot product ────────────────────────────────────────────────────────────
sp_dot = (measure(lambda: sum(a*b for a,b in zip(v_l,w_l))) /
          measure(lambda: np.dot(v_np, w_np)))

# ── 3. Matrix–vector product ──────────────────────────────────────────────────
sp_mv = (measure(lambda: [sum(A_l[i][j]*v_l[j] for j in range(N)) for i in range(N)], repeats=3) /
         measure(lambda: A_np @ v_np))

print(f"element-wise x²  : {sp_sq:.1f}×  speedup")
print(f"dot product      : {sp_dot:.1f}×  speedup")
print(f"mat-vec product  : {sp_mv:.1f}×  speedup")
````
